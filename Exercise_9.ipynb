```python
data = pd.read_csv("video_game_reviews.csv")
dataFrame = pd.DataFrame(data)
dataFrame
```

```python
import matplotlib.pyplot as plt
```

```python
x = dataFrame["Release Year"]
y = dataFrame["User Rating"]

plt.plot(x, y)
plt.xticks(fontsize=6, rotation='vertical')
plt.title("Price Overtime")
plt.xlabel("Date of Release")
plt.ylabel("Price (USD)")
plt.show()
```

```python
x = dataFrame["Release Year"]
y = dataFrame["User Rating"]

plt.plot(x, y, 'o:b', color ='red', marker='o', mfc='red',linewidth=3)
plt.xticks(fontsize=6, rotation='vertical')
plt.title("Price Overtime")
plt.xlabel("Year")
plt.ylabel("Price (USD)")
plt.show()
```

```python
sales = dataFrame.groupby(['Graphics Quality', 'Genre'])['User Rating'].sum().unstack()

for brand in sales.columns:
    plt.plot(sales.index, sales[brand], marker='o', label=brand)

plt.legend(bbox_to_anchor=(1,1))
plt.title("Product Sales")
plt.xlabel("Devices")
plt.xticks(fontsize=6, rotation='vertical')
plt.ylabel("Price")
plt.show()
```

```python
xbar = dataFrame["Genre"]
ybar = dataFrame["User Rating"]

plt.bar(xbar, ybar)
plt.xticks(fontsize=8, rotation='vertical')
plt.title("Brand Sales")
plt.xlabel("Brand")
plt.ylabel("Price (USD)")
plt.show()
```

```python
xhist = dataFrame["Developer"]

plt.hist(xhist)
plt.xticks(fontsize=8, rotation='vertical')
plt.title("Country of Origin")
plt.xlabel("Country")
plt.ylabel("Frequency")
plt.show()
```
