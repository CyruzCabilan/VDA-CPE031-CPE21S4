```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.array([0.5, 1, 1.3, 1.6, 2, 2.2, 2.5, 3.1, 3.9, 4.4])
y = np.array([1.9, 1.5, 1.2, 1, 0.8, 0.78, 0.65, 0.46, 0.3, 0.23])

slope, intercept = np.polyfit(x, y, 1)

y_pred = slope * x + intercept

plt.scatter(x, y)
plt.plot(x, y_pred, color='green')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Data Set')
plt.show()
```
