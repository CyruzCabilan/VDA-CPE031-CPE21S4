```python
import pandas as pd

df = pd.read_csv('heart_2020_cleaned.csv')

print(df)
```

```python
import matplotlib.pyplot as plt
# Question 1: Is there a correlation between age and physical health?

plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['PhysicalHealth'], alpha=0.5)
plt.title('Correlation between Age and Physical Health')
plt.xlabel('Age')
plt.ylabel('Physical Health (Days)')
plt.grid(True)
plt.show()
```

```python
# Question 2: Is there a difference in physical health between smokers and non-smokers?
smoker_physical_health = df.groupby('Smoking')['PhysicalHealth'].mean()

plt.figure(figsize=(8, 6))
plt.bar(smoker_physical_health.index, smoker_physical_health.values, color=['Green', 'red'])
plt.title('Average Physical Health by Smoking Status')
plt.xlabel('Smoker')
plt.ylabel('Average Physical Health (Days)')
plt.grid(axis='y')
plt.show()
```
