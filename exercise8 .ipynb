#**Introduction to Matplotlib**


```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4]
plt.hist(data)
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values = [5, 7, 3]
plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [[1.5]*10 + [2]*10 + [3]*10]

plt.boxplot(data)

plt.title("Box Plot Example")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C']
sizes = [40, 30, 30]

plt.pie(sizes, labels=labels)
plt.title("Pie Chart Example")
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('Sport car price.csv')

print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot: Engine Size vs Horsepower
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Engine Size (L)', y='Horsepower', hue='Car Make', data=df, palette='viridis')

plt.title('Engine Size vs Horsepower by Car Make')
plt.xlabel('Engine Size (L)')
plt.ylabel('Horsepower')
plt.legend(title='Car Make', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Box plot: Torque by Car Make
plt.figure(figsize=(12, 8))
sns.boxplot(x='Car Make', y='Torque (lb-ft)', data=df, palette='Set2')

plt.title('Distribution of Torque by Car Make')
plt.xlabel('Car Make')
plt.ylabel('Torque (lb-ft)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
