```python

import matplotlib.pyplot as plt
import numpy as np

#Coordinates
x = np.array([0,6])
y = np.array([0,250])

plt.plot(x, y) #or you could only plot either x or y only
plt.show() #shows the plotted graph
```

```python
arr1 = np.array([3,8,1,10])
plt.plot(arr1, marker='o') #Adding marker means the plot will make lines along with the rings
plt.show()
```

```python
plt.plot(arr1, marker='x') #'x' means x marks instead of rings
plt.show()
```

```python
plt.plot(arr1, 'o:b', ms=20) # 'o:b' means having ring markers and broken lines, and ms increases the size of the markers
plt.show()
```

```python
plt.plot(arr1, 'o:b', ms=10, mec='r') # Adding mec adds colored borders to the markers. In this case, 'r' means red
plt.show()
```

```python
plt.plot(arr1, 'o:b', ms=25, mec='green',mfc='red') #mfc changes the color of the markers
plt.show()
```

```python
plt.plot(arr1, linestyle='dashed') #linestyle also changes the linestyle duhh
plt.show()
```

```python
plt.plot(arr1, linestyle='--') #another way of saying dashed
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot: Engine Size vs Horsepower
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Engine Size (L)', y='Horsepower', hue='Car Make', data=df, palette='viridis')

plt.title('Engine Size vs Horsepower by Car Make')
plt.xlabel('Engine Size (L)')
plt.ylabel('Horsepower')
plt.legend(title='Car Make', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame (replace this with your actual DataFrame)
data = {
    'Engine Size (L)': [1.6, 2.0, 2.4, 3.0, 3.5, 4.0],
    'Horsepower': [100, 130, 150, 200, 250, 300]
}
df = pd.DataFrame(data)

# Create a scatter plot of engine size vs horsepower
plt.scatter(df['Engine Size (L)'], df['Horsepower'], color='blue')

# Set title and labels
plt.title('Relationship Between Engine Size and Horsepower')
plt.xlabel('Engine Size (L)')
plt.ylabel('Horsepower')

# Add a trend line
z = np.polyfit(df['Engine Size (L)'], df['Horsepower'], 1)
p = np.poly1d(z)
plt.plot(df['Engine Size (L)'], p(df['Engine Size (L)']), "r--", label='Trend Line')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame (replace this with your actual DataFrame)
data = {
    'Price (in USD)': [20000, 25000, 30000, 15000, 18000, 40000, 35000, 22000, 27000, 32000],
    'Horsepower': [100, 130, 150, 200, 250, 300, 280, 220, 180, 160]
}
df = pd.DataFrame(data)

# Create a figure with two subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Create a histogram of price values on the first subplot
ax[0].hist(df['Price (in USD)'], bins=10, edgecolor='black')
ax[0].set_title('Distribution of Prices Across Different Car Models')
ax[0].set_xlabel('Price (in USD)')
ax[0].set_ylabel('Frequency')

# Create a histogram of horsepower values on the second subplot
ax[1].hist(df['Horsepower'], bins=10, edgecolor='black')
ax[1].set_title('Distribution of Horsepower Across Different Car Models')
ax[1].set_xlabel('Horsepower')
ax[1].set_ylabel('Frequency')

# Layout so plots do not overlap
fig.tight_layout()

# Show the plot
plt.show()
```
